```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data preprocessing

```python
current_folder = os.getcwd()
file_path = os.path.join(current_folder,'..' ,'resource', 'RSMC_Best_Track_Data.csv')
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
new_order = [
    'International number ID',
    'Name of the storm',
    'Time of analysis',
    'Grade',
    'Central pressure',
    'Maximum sustained wind speed',
    'Direction of the longest radius of 50kt winds or greater',
    'The longest radius of 50kt winds or greater',
    'The shortest radius of 50kt winds or greater',
    'Direction of the longest radius of 30kt winds or greater',
    'The longest radius of 30kt winds or greater',
    'The shortest radius of 30kt winds or greater',
    'Indicator of landfall or passage',
    'Latitude of the center',
    'Longitude of the center'
]

df = df[new_order]
df
```

```python
df.info()
```

```python
# 加上next的经纬度

# 添加两列并初始化为NaN
df['Latitude of the center_next'] = float('nan')
df['Longitude of the center_next'] = float('nan')

# 遍历每个国际编号（International number ID）
for international_id in df['International number ID'].unique():
    # 获取该国际编号的所有行
    rows = df[df['International number ID'] == international_id]
    
    # 遍历每行，设置Latitude of the center_next 和 Longitude of the center_next
    for i in range(len(rows) - 1):
        df.at[rows.index[i], 'Latitude of the center_next'] = rows.at[rows.index[i + 1], 'Latitude of the center']
        df.at[rows.index[i], 'Longitude of the center_next'] = rows.at[rows.index[i + 1], 'Longitude of the center']

# 处理每组国际编号的最后一行
for international_id in df['International number ID'].unique():
    rows = df[df['International number ID'] == international_id]
    last_index = rows.index[-1]
    
    # 处理最后一行数据
    # df.at[last_index, 'Latitude of the center_next'] = df.at[last_index, 'Latitude of the center']
    # df.at[last_index, 'Longitude of the center_next'] = df.at[last_index, 'Longitude of the center']
    df.drop(last_index, inplace=True)

df.info()
```

```python
df.info()
```

```python
columns_to_check = [
    'Direction of the longest radius of 50kt winds or greater',
    'The longest radius of 50kt winds or greater',
    'The shortest radius of 50kt winds or greater',
    'Direction of the longest radius of 30kt winds or greater',
    'The longest radius of 30kt winds or greater',
    'The shortest radius of 30kt winds or greater'
]
df = df.dropna(subset=columns_to_check)
df.info()
```

### Make Correlation Matrix

```python
df_corr = df.copy()
```

```python
direction_30_mapping = {'(symmetric circle)': 1, 'Northeast (NE)': 6, 'South (S)': 3,
       'East (E)': 5, 'Southeast (SE)': 9, 'West (W)': 4, 'North (N)': 2,
       'Northwest (NW)': 6, 'Southwest (SW)': 8}
df_corr['Direction of the longest radius of 30kt winds or greater'] = df_corr['Direction of the longest radius of 30kt winds or greater'].map(direction_30_mapping)


direction_50_mapping = {'(symmetric circle)': 1, 'Northeast (NE)': 6, 'South (S)': 3,
       'East (E)': 5, 'Southeast (SE)': 9, 'West (W)': 4, 'North (N)': 2,
       'Northwest (NW)': 6, 'Southwest (SW)': 8, 'No direction (Longest radius of 50kt winds is 0)': 9}
df_corr['Direction of the longest radius of 50kt winds or greater'] = df_corr['Direction of the longest radius of 50kt winds or greater'].map(direction_50_mapping)


indicator_mapping = {'#': 1, ' ': 0}
df_corr['Indicator of landfall or passage'] = df_corr['Indicator of landfall or passage'].map(indicator_mapping)
```

```python
df_corr = df_corr.drop(['International number ID', 'Name of the storm' ,'Grade', 'Time of analysis'], axis=1)
```

```python
corr = df_corr.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt='.2f')
plt.show()
```

### One-hot-encoded

```python
data = df.drop(['International number ID', 'Name of the storm', 'Grade','Indicator of landfall or passage'], axis=1)
```

```python
columns_to_check = [
    'Direction of the longest radius of 50kt winds or greater',
    'The longest radius of 50kt winds or greater',
    'The shortest radius of 50kt winds or greater',
    'Direction of the longest radius of 30kt winds or greater',
    'The longest radius of 30kt winds or greater',
    'The shortest radius of 30kt winds or greater'
]
df = df.dropna(subset=columns_to_check)
df.info()
```

```python
one_hot_encoded = pd.get_dummies(data['Direction of the longest radius of 50kt winds or greater'],prefix = 'Direction of the longest radius of 50kt winds or greater', dtype = int)
data = pd.concat([data, one_hot_encoded], axis = 1)

one_hot_encoded2 = pd.get_dummies(data['Direction of the longest radius of 30kt winds or greater'], prefix= 'Direction of the longest radius of 30kt winds or greater', dtype = int)
data = pd.concat([data, one_hot_encoded2], axis = 1)


data = data.drop(['Direction of the longest radius of 50kt winds or greater', 'Direction of the longest radius of 30kt winds or greater'], axis=1)
data.tail()
```

```python
data.info()
```

```python
data['Time of analysis'] = pd.to_datetime(data['Time of analysis'])
data['Time of analysis'] = data['Time of analysis'].apply(lambda x: x.timestamp())
data['Time of analysis']
```

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = data.columns[:]

# 复制 DataFrame 以免修改原始数据
df_normalized = data.copy()

# 实例化 MinMaxScaler
scaler = MinMaxScaler()

# 对选择的列进行归一化
df_normalized[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# 打印处理后的新 DataFrame 的信息
df_normalized.tail()
```

### MODEL BUILDING!!!!

```python
split_index = int(0.7 * len(data))

# 分割DataFrame
train_set = data.iloc[:split_index, :]
test_set = data.iloc[split_index:, :]

train_x = train_set.drop(["Latitude of the center_next", "Longitude of the center_next"], axis=1)
train_y = train_set.loc[:, ["Latitude of the center_next", "Longitude of the center_next"]]
test_x = test_set.drop(["Latitude of the center_next", "Longitude of the center_next"], axis=1)
test_y = test_set.loc[:, ["Latitude of the center_next", "Longitude of the center_next"]]


x_train = train_x.copy()
y_train = train_y.copy()
x_test = test_x.copy()
y_test = test_y.copy()

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
train_x = np.reshape(train_x, (train_x.shape[0], 1, train_x.shape[1]))
test_x = np.reshape(test_x, (test_x.shape[0], 1, test_x.shape[1]))

print(train_x.shape)
print(test_x.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, BatchNormalization, Dropout,Dense
from keras.optimizers import Adam
```

```python
model = Sequential()

model.add(LSTM(128, return_sequences=True, input_shape=(1, 28)))
model.add(BatchNormalization())
#model.add(Dropout(0.2))
model.add(LSTM(64, return_sequences=True))
model.add(LSTM(32, return_sequences=False))
#model.add(Dropout(0.2))
model.add(Dense(32))
model.add(Dense(2))
```

```python
optimizer = Adam(learning_rate=0.00001)
model.compile(loss='mean_absolute_error', optimizer=optimizer)
history = model.fit(train_x, train_y, epochs=20, batch_size=30, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
testPredict = model.predict(test_x)
# testPredict = MinMaxScaler.inverse_transform(testPredict)
testPredict
```

```python
last_20_rows = test_y.tail(20)
last_20_rows_numpy_array = testPredict[20:]

# 绘制图表
plt.scatter(last_20_rows['Longitude of the center_next'], last_20_rows['Latitude of the center_next'])
plt.scatter(last_20_rows_numpy_array[:, 0], last_20_rows_numpy_array[:, 1], label='numpy_array')

# 设置图表标题和轴标签
plt.title('Track Sample')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 显示图表
plt.show()
```
