```python
import os
import pandas as pd
import numpy as np
```

```python
current_folder = os.getcwd()
file_path = os.path.join(current_folder,'..' ,'resource', 'RSMC_Best_Track_Data.csv')
df = pd.read_csv(file_path)
columns_to_check = [
        'Direction of the longest radius of 50kt winds or greater',
        'The longest radius of 50kt winds or greater',
        'The shortest radius of 50kt winds or greater',
        'Direction of the longest radius of 30kt winds or greater',
        'The longest radius of 30kt winds or greater',
        'The shortest radius of 30kt winds or greater'
    ]
df = df.dropna(subset=columns_to_check)

to_file_path = os.path.join(current_folder, '..', 'resource', 'RSMC_Best_Track_Data_no_null.csv')

df.to_csv(to_file_path)
```

```python
df = df.sort_values(by='International number ID')

# 找到所有可能的'International number ID'
unique_ids = df['International number ID'].unique()

# 从中随机选择一个ID
selected_id = np.random.choice(unique_ids, size=1)[0]

# 获取该ID对应的所有行
selected_rows = df[df['International number ID'] == selected_id]

# 随机选择10行
if len(selected_rows) >= 10:
    random_10_rows = selected_rows.sample(n=10)


random_10_rows
```

```python
initial_coordinates = random_10_rows[["Longitude of the center", "Latitude of the center"]]
for i in range(len(initial_coordinates)):
    print(initial_coordinates.iloc[i])
```
